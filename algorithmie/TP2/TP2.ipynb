# Listes chaînées

Ci-dessous un objet minimal pour représenter les listes chaînées en python

```python
class Cell:
    
    def __init__(self, valeur):
        self.val = valeur
        self.next = None
    
    def __repr__(self):
        return str(self.val)

class List:
    
    def __init__(self):
        self.head = None
    
    def __repr__(self):
        s = "["
        c = self.head
        while not c is None:
            s += str(c)
            if c.next is not None:
                s+=" -> "
            c = c.next
        s+="]"
        return s
```

Vous n'avez pas besoin de comprendre entièrement ce code, ni même de maîtriser la programmation objet pour la suite du TP. Voilà comment vous pouvez l'utiliser.

Création d'une cellule :

```python
c1 = Cell(4)
c2 = Cell(8)
print(c1)
print(c2)
```

Accès aux différents champs :

```python
print(c1.val)
```

```python
c1.next = c2
```

```python
c2.next is None
```

Création de la liste

```python
L = List()
print(L)
```

Comme j'ai accroché c2 à c1, il me suffit de modifier *L.head* pour obtenir la liste contenant c1 puis c2.

```python
def add(L,v):
    """
    Ajoute une cellule contenant la valeur v en tête de liste
    Input :
        - L, une liste chaînée
        - v, une valeur
    La fonction doit créer la cellule contenant la valeur v et l'ajouter en tête de liste.
    """
    c = Cell(v)
    if L.head != None:
        c.next = L.head
    L.head = c
```

```python
LTest = List()
add(LTest,3)
add(LTest,2)
add(LTest,1)
assert str(LTest)=="[1 -> 2 -> 3]"
```

```python
def pop(L):
    """
    Supprime la cellule en tête de liste
    Input : L, une liste chaînée
    Output : la valeur de la cellule supprimée. Si la liste est vide, on renvoie None et on ne fait rien.
    """
    if L.head != None:
        val = L.head.val
        if L.head.next != None:
            L.head = L.head.next
            return val
        else:
            L.head = None
            return val
    return None
```

```python
LTest = List()
add(LTest,3)
add(LTest,2)
add(LTest,1)
assert pop(LTest)==1
assert pop(LTest)==2
assert pop(LTest)==3
assert pop(LTest) is None
assert str(LTest)=="[]"
```

```python
def toList(l: list):
    """
    Transforme une liste classique Python en lîste chaînée
    Input :
        - une liste Python (structure de tableau dynamique)
    Output: la liste chaînée correspondante
    """
    lst = List()
    for i in range(len(l) -1, -1, -1):
        add(lst, l[i])
    return lst
```

```python
assert str(toList([1,2,3,4]))=="[1 -> 2 -> 3 -> 4]"
assert str(toList([]))=="[]"
assert str(toList([5,2,4,1,3]))=="[5 -> 2 -> 4 -> 1 -> 3]"
```

```python
def remove(L, v):
    """
    Supprime la premiere occurence de v dans L
    Input :
        - L, une liste chaînée
        - v, une valeur
    Output : True si une valeur a été supprimée et False sinon
    """
    # Si la liste est vide
    if L.head is None:
        return False

    # Cas où la valeur est en tête
    if L.head.val == v:
        L.head = L.head.next
        return True

    # Recherche dans le reste de la liste
    c = L.head
    while c.next is not None:
        if c.next.val == v:
            c.next = c.next.next
            return True
        c = c.next

    return False
```

```python
LTest = List()
add(LTest,3)
add(LTest,2)
add(LTest,1)
add(LTest,2)
assert remove(LTest,1)
assert remove(LTest,2)
assert remove(LTest,3)
assert not remove(LTest,1)
assert remove(LTest,2)
assert str(LTest)=="[]"
```

```python
def insertSorted(L,v):
    """
    Ajoute une cellule contenant la valeur v dans une liste triée, en maintenant le tri.
    Input :
        - L, une liste chaînée dont les valeurs sont triées par ordre croissant
        - v, une valeur
    La fonction doit créer la cellule contenant la valeur v et l'ajouter à la bonne position dans L.
    """
    nouv = Cell(v)

    # Insertion en tête (liste vide ou plus petite valeur)
    if L.head is None or v <= L.head.val:
        nouv.next = L.head
        L.head = nouv
        return
    
    # Recherche de la position d'insertion
    c = L.head
    while c.next is not None and c.next.val < v:
        c = c.next

    # Insertion après c
    nouv.next = c.next
    c.next = nouv
```

```python
LTest = List()
insertSorted(LTest,6)
insertSorted(LTest,3)
insertSorted(LTest,10)
insertSorted(LTest,4)
insertSorted(LTest,4)
insertSorted(LTest,1)
assert str(LTest)=="[1 -> 3 -> 4 -> 4 -> 6 -> 10]"
```

Maintenant que l'on sait insérer dans une liste triée, on peut implanter un algorithme de tri sur les listes : il suffit de récupérer les cellules une à une, puis de les insérer dans une nouvelle liste. 

```python
def sortList(L):
    """
    Trie la liste L donnée en entrée
    Input :
        - L, une liste chaînée 
    """
    nouv = List()
    while L.head != None:
        insertSorted(nouv, pop(L))
    L.head = nouv.head
```

```python
LTest = List()
add(LTest,3)
add(LTest,2)
add(LTest,1)
add(LTest,2)
sortList(LTest)
assert str(LTest) == "[1 -> 2 -> 2 -> 3]"
```
