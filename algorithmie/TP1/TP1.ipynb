```python
def complexiteN(n):
    i = 0
    while i < n:
        i += 1
```

```python
def complexiteN2(n):
    i, j = 0, 0
    while  i < n:
        while j < n:
            j += 1
        i += 1
        j = 0
```

```python
def complexiteN3(n):
    i, j, k = 0, 0, 0
    while  i < n:
        while j < n:
            while k < n:
                k += 1
            j += 1
            k = 0
        i += 1
        j = 0
```

## Recherche dichotomique

Compléter les deux fonctions suivantes qui effectuent une recherche dans un tableau. Dans le second cas, on suppose que le tableau est trié et on effectue donc une recherche dichotomique.

```python
def recherche(l, v):
    """
    Recherche classique dans un tableau
    Input :
        - l, une liste d'entiers
        - v, une valeur
    Output : True si v est dans le tableau, False sinon
    """
    for x in l:
        if x == v:
            return True
    return False
```

```python
assert recherche([5,2,4,2,6,4,6], 4)
assert recherche([5,2,4,2,6,4,6], 5)
assert recherche([5,2,4,2,6,4,6], 2)
assert recherche([5,2,4,2,6,4,6], 6)
assert not recherche([5,2,4,2,6,4,6], 10)
```

```python
def rechercheDicho(l, v):
    """
    Recherche dichotomique dans un tableau trié
    Input :
        - l, une liste d'entiers
        - v, une valeur
    Output : True si v est dans le tableau, False sinon
    """
    debut = 0
    fin = len(l) - 1
    while debut <= fin:
        m = (debut + fin) // 2
        if l[m] == v:
            return True
        elif l[m] < v:
            debut = m + 1
        else:
            fin = m - 1
    return False
```

```python
assert rechercheDicho([2,2,4,4,5,6], 4)
assert rechercheDicho([2,2,4,4,5,6], 5)
assert rechercheDicho([2,2,4,4,5,6], 2)
assert rechercheDicho([2,2,4,4,5,6], 6)
assert not rechercheDicho([2,2,4,4,5,6], 10)
```

Comme Python possède la fonction *sort*, on se dit que peut-être, on pourrait modifier la recherche dans les tableaux non triés pour d'abord trier le tableau puis effectuer une recherche dichotomique.

**Complétez la fonction suivante**

```python
def recherchePresqueDicho(l, v):
    """
    Trie le tableau en entrée puis effectue une recherche dichotomique de la valeur v
    Input :
        - l, une liste d'entiers
        - v, une valeur
    Output : True si v est dans le tableau, False sinon
    """
    # écrire le code
    l.sort()
    return rechercheDicho(l, v)
```

```python
assert recherchePresqueDicho([5,2,4,2,6,4,6], 4)
assert recherchePresqueDicho([5,2,4,2,6,4,6], 5)
assert recherchePresqueDicho([5,2,4,2,6,4,6], 2)
assert recherchePresqueDicho([5,2,4,2,6,4,6], 6)
assert not recherchePresqueDicho([5,2,4,2,6,4,6], 10)
```

Comparez les performances de *recherche* et *recherchePresqueDicho*

```python
debut1 = time.perf_counter()
recherche(t1, valeur)
fin1 = time.perf_counter()

print(f"Temps d'exécution recherche (100): {fin1 - debut1:.6f} s")

debut2 = time.perf_counter()
recherchePresqueDicho(t1, valeur)
fin2 = time.perf_counter()

print(f"Temps d'exécution presqueDicho (100): {fin2 - debut2:.6f} s")

speedup = (fin1 - debut1) / (fin2 - debut2) if (fin2 - debut2) > 0 else float("inf")
print(f"Speedup : {speedup:.2f}x")
```
